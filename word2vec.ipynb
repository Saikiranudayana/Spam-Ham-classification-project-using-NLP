```python
df = pd.read_csv(
    'spam.csv',
    encoding='latin-1',
    usecols=[0, 1],             # only read the first 2 columns
    names=['label', 'message'], # rename them right away
    skiprows=1                  # skip the original header row
)
```

```python
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
import re
corpus = []
for i in range(0,len(df)):
    review=re.sub('[^a-zA-Z]',' ',df['message'][i])
    review= review.lower()
    review=review.split()
    review = [lemmatizer.lemmatize(word) for word in review ]
    review= ' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus,df['message'])  ]
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for setn in sent_token:
        words.append(simple_preprocess(sent))
```

```python
import numpy as np
def avg_word2vec(doc):
    # remove out-of-vocabulary words 
    #sent = [words for words in doc if words in model.wv.index_to_key]
    #print(sent)
    
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key])
```

```python
import pandas as pd
import numpy as np

cf_list = []  # Store DataFrames in a list

for i in range(len(x)):
    row = np.array(x[i])  # Convert to NumPy array

    # Ensure the row is a 1D array of length 100
    if row.ndim == 1:  
        row = row.reshape(1, -1)  # Convert to a single-row 2D array

    elif row.shape[0] == 100 and row.shape[1] == 1:  
        row = row.T  # Transpose if it's (100,1)

    # Convert to DataFrame and append
    row_df = pd.DataFrame(row)
    cf_list.append(row_df)

# Concatenate all rows into a DataFrame
cf = pd.concat(cf_list, ignore_index=True)

print(f"Final shape: {cf.shape}")  # Should be (5569, 100)
```
